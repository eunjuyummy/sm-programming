```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re # 정규표현의 처리를 실행하는 라이브러
```

```python
movies_data=pd.read_csv('/content/drive/MyDrive/머신러닝/input/ml-latest-small/movies.csv')
ratings_data=pd.read_csv('/content/drive/MyDrive/머신러닝/input/ml-latest-small/ratings.csv')
```

```python
movies_data
```

```python
ratings_data
```

```python
ratings_data['movieId'].value_counts().sort_index()
```

```python
# 동일한 영화 ID의 평균 평점을 계산
average_ratings = ratings_data.groupby('movieId')['rating'].mean()

# 결과 출력
print(average_ratings)
```

```python
new_movie_data = movies_data
new_movie_data
```

```python
new_movie_data['parsed_genres'] = new_movie_data['genres'].apply(lambda x: re.split(r'\|', x)[0])
new_movie_data = new_movie_data.merge(average_ratings, on='movieId')
new_movie_data
```

```python
new_movie_data = new_movie_data.drop(new_movie_data[new_movie_data['rating'] < 3.0].index)
new_movie_data
```

```python
new_movie_data = new_movie_data[new_movie_data['parsed_genres'] != '(no genres listed)']
new_movie_data
```

```python
# CSV 파일로 저장
new_movie_data.to_csv('movie_data.csv', index=False)
```
